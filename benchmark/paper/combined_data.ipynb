```python
import pandas as pd
import json
import os
from collections import OrderedDict

# from .agent_action_regex import is_action

# Function to convert JSON-like strings to nested dictionaries
def nested_json(x):
    if pd.notna(x):
        d = json.loads(x)
        if "content" in d and isinstance(d["content"], str):
            try:
                d["content"] = json.loads(d["content"])
            except json.JSONDecodeError:
                pass
        return d
    return x

challenge = "TestThreeSum"
agent_array = ['beebot'] # df['agent'].unique()
request_type = 'request' # 'request' or 'response'

# Loop through unique agents
for agent in agent_array:
    selected_df = df.loc[(df['agent'] == agent) & (df['challenge'] == challenge)]
    
    master_response_dict = OrderedDict()
    master_response_nested_dict = OrderedDict()
    master_actions_dict = OrderedDict()

    # Group by 'benchmark_start_time'
    grouped_df = selected_df.groupby('benchmark_start_time')

    for timestamp, group in grouped_df:
        response_dict = OrderedDict()
        response_nested_dict = OrderedDict()
        actions_dict = OrderedDict()
        
        total_rows = len(group)
        
        for i, (_, row) in enumerate(group.iterrows()):
            if request_type == 'response':
                response = json.loads(row['response'])
                response_nested = nested_json(row['response'])
                
                response_nested_dict[str(total_rows-i)] = response_nested
            else:
                response = row['request']
            
            response_dict[str(total_rows-i)] = response  # Starting from 1 as you mentioned
            
            if request_type == 'response' and is_action_agent(response, agent, challenge, response):
                actions_dict[str(total_rows-i)] = response
        
        if request_type == 'response':
            response_nested_dict = OrderedDict(reversed(list(response_nested_dict.items())))
            actions_dict = OrderedDict(reversed(list(actions_dict.items())))
            
            master_response_nested_dict[str(timestamp)] = response_nested_dict
            master_actions_dict[str(timestamp)] = actions_dict
            
        # Reverse the order within each timestamp group
        response_dict = OrderedDict(reversed(list(response_dict.items())))

        # Add the timestamp as a key to the master dictionary
        master_response_dict[str(timestamp)] = response_dict
        
    
    # Create directories
    os.makedirs(f'{challenge}', exist_ok=True)
    os.makedirs(f'{challenge}/{agent}', exist_ok=True)
    
    # Write to a JSON file specific to the agent
    with open(f'{challenge}/{agent}/selected_logs_{request_type}.json', 'w') as f:
        json.dump(master_response_dict, f, indent=4)
         
    if request_type == 'response':
        with open(f'{challenge}/{agent}/flagged_actions.json', 'w') as f:
            json.dump(master_actions_dict, f, indent=4)
            
        with open(f'{challenge}/{agent}/selected_logs_nested_response.json', 'w') as f:
            json.dump(master_response_nested_dict, f, indent=4)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Filter out agents to ignore
filtered_df = df[~df['agent'].isin(agents_to_ignore)]

# Group by 'agent' and 'challenge' to calculate the mean of 'success'
agg_df = filtered_df.groupby(['agent', 'challenge']).agg({
    'success': 'mean'
}).reset_index()

# Remove challenges with no successful data (where average success is 0)
agg_df = agg_df[agg_df.groupby('challenge')['success'].transform('mean') > 0]

# Remove challenges to ignore
agg_df = agg_df[~agg_df['challenge'].isin(challenges_to_ignore)]

# Get unique challenges
unique_challenges = agg_df['challenge'].unique()
mid_idx = len(unique_challenges) // 2

# Divide the challenges into two halves
first_half_challenges = unique_challenges[:mid_idx]
second_half_challenges = unique_challenges[mid_idx:]

# Function to plot the graph
def plot_graph(challenges, graph_num):
    fig, ax = plt.subplots(figsize=(15, 10))

    barWidth = 0.15
    r1 = np.arange(len(challenges))
    r_list = [r1]

    for idx, agent in enumerate(agg_df['agent'].unique()):
        if idx > 0:
            r_list.append([x + barWidth for x in r_list[-1]])

        subset = agg_df[agg_df['agent'] == agent]
        subset = subset[subset['challenge'].isin(challenges)]
        subset = subset.set_index('challenge').reindex(challenges).reset_index()
        ax.bar(r_list[-1], subset['success'], color='C'+str(idx), width=barWidth, edgecolor='grey', label=agent)

    ax.set_xticks([r + barWidth for r in range(len(challenges))])
    ax.set_xticklabels(challenges, rotation=45)
    ax.set_xlabel('Challenges', fontweight='bold')
    plt.title(f'Average Success % per Agent for Each Challenge (Graph {graph_num})')
    plt.ylabel('Average Success %')
    plt.legend()

    plt.show()

# Plot first half
plot_graph(first_half_challenges, 1)

# Plot second half
plot_graph(second_half_challenges, 2)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Filter out agents to ignore
filtered_df = df[~df['agent'].isin(agents_to_ignore)]

# Group by 'agent', 'challenge', and 'benchmark_start_time' to calculate sum of successes and count of attempts
agg_df = filtered_df.groupby(['agent', 'challenge', 'benchmark_start_time']).agg({
    'success': 'sum',
    'attempted': 'count'
}).reset_index()

# Calculate 'successes per attempt'
agg_df['success_per_attempt'] = agg_df['success'] / agg_df['attempted']

# Remove challenges with no successful data (where average success is 0)
agg_df = agg_df[agg_df.groupby('challenge')['success'].transform('mean') > 0]

# Remove challenges to ignore
agg_df = agg_df[~agg_df['challenge'].isin(challenges_to_ignore)]

# Get unique challenges
unique_challenges = agg_df['challenge'].unique()
mid_idx = len(unique_challenges) // 2

# Divide the challenges into two halves
first_half_challenges = unique_challenges[:mid_idx]
second_half_challenges = unique_challenges[mid_idx:]

# Function to plot the graph
def plot_graph(challenges, graph_num):
    fig, ax = plt.subplots(figsize=(15, 10))

    barWidth = 0.15
    r1 = np.arange(len(challenges))
    r_list = [r1]

    for idx, agent in enumerate(agg_df['agent'].unique()):
        if idx > 0:
            r_list.append([x + barWidth for x in r_list[-1]])

        subset = agg_df[agg_df['agent'] == agent]
        subset = subset[subset['challenge'].isin(challenges)]
        subset = subset.groupby('challenge')['success_per_attempt'].mean().reset_index()
        subset = subset.set_index('challenge').reindex(challenges).reset_index()
        ax.bar(r_list[-1], subset['success_per_attempt'], color='C'+str(idx), width=barWidth, edgecolor='grey', label=agent)

    ax.set_xticks([r + barWidth for r in range(len(challenges))])
    ax.set_xticklabels(challenges, rotation=45)
    ax.set_xlabel('Challenges', fontweight='bold')
    plt.title(f'Average Successes per Attempt per Agent for Each Challenge (Graph {graph_num})')
    plt.ylabel('Successes per Attempt')
    plt.legend()

    plt.show()

# Plot first half
plot_graph(first_half_challenges, 1)

# Plot second half
plot_graph(second_half_challenges, 2)
```
